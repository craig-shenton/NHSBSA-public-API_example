```python
import urllib.request
import requests
import pandas as pd
offset = list(range(0, 5000, 1000))
base_url = "https://opendata.nhsbsa.net/api/3/action/datastore_search?"
dataset_id = "&resource_id=EPD_"
dates = ['202105', '202106', '202107', '202108', '202109', '202110', '202111', '202112', '202201', '202202', '202203', '202204']
query1 = '&fields=BNF_CHEMICAL_SUBSTANCE+STP_NAME+QUANTITY+TOTAL_QUANTITY'
limit = '&limit=1000'
search = '&q={%22BNF_CHEMICAL_SUBSTANCE%22:%20%220501.%22}' # match 5.1: all antibacterial drugs https://openprescribing.net/bnf/0501/
month_list = []

for date in dates:
  for offset_value in offset:
    url = f'{base_url}offset={offset_value}{dataset_id}{date}{query1}{limit}{search}'
    json_file = requests.get(url).json()
    month_data = pd.json_normalize(json_file['result']['records'])
    month_list.append(month_data)
  print(url)
df = pd.concat(month_list)

# https://opendata.nhsbsa.net/api/3/action/datastore_search?q=jones&resource_id=EPD_202104
```
